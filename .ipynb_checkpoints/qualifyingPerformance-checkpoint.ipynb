```python
import pandas as pd
import fastf1 as ff1
from fastf1 import plotting
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb
import matplotlib
import numpy as np
from fastf1 import plotting
import seaborn as sns
```

```python
ff1.Cache.enable_cache('cache')
plotting.setup_mpl()
```

```python
year = 2022
schedule = ff1.get_event_schedule(year)
```

```python
quali_laps = []
quali_sessions = []
for _, event in schedule.iterrows():
    session = ff1.get_session(year, event['Location'], 'Q')
    session.load()
    quali_sessions.append(session)
    quali_laps.append(session.laps)
```

```python
quali_laps_copy = quali_laps[0].copy()
quali_laps_copy['LapTime'] = quali_laps_copy['LapTime'].dt.total_seconds()
fastest_laps = quali_laps_copy.groupby(['Driver', 'Team'])['LapTime'].min()
fastest_laps = fastest_laps.reset_index()
fastest_laps['Rank'] = fastest_laps['LapTime'].rank().astype(int)
fastest_laps = fastest_laps.sort_values('Rank')
```

```python
ranking_qualifiers = []
for laps in quali_laps:
    quali_laps_copy = laps.copy()
    quali_laps_copy['LapTime'] = quali_laps_copy['LapTime'].dt.total_seconds()
    fastest_laps = quali_laps_copy.groupby(['Driver', 'Team'])['LapTime'].min()
    fastest_laps = fastest_laps.reset_index()
    fastest_laps['Rank'] = fastest_laps['LapTime'].fillna(100000000).rank().astype(int)
    fastest_laps = fastest_laps.sort_values('Rank')
    ranking_qualifiers.append(fastest_laps)
```

```python
# Collect all unique drivers and their teams
drivers_teams = {}
for df in ranking_qualifiers:
    for _, row in df.iterrows():
        drivers_teams[row['Driver']] = row['Team']

# Plotting ranking over qualifiers for each driver
plt.figure(figsize=(20, 10))

for driver, team in drivers_teams.items():
    ranks = []
    for i, df in enumerate(ranking_qualifiers):
        rank = df[df['Driver'] == driver]['Rank'].values
        if len(rank) > 0:
            ranks.append(rank[0])
        else:
            ranks.append(np.nan)  # Driver didn't participate in this qualifier
    plt.plot(range(1, len(ranking_qualifiers) + 1), ranks, label=driver, color=plotting.team_color(team))

plt.title('Driver Ranking over Qualifiers')
plt.xlabel('Qualifier Number')
plt.ylabel('Rank')
plt.legend()
plt.gca().invert_yaxis()  # So that rank 1 is at the top
plt.show()
```
